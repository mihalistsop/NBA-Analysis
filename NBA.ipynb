<h1>Exploring NBA Data 

```python
#Importing Essential Libraries
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
pd.set_option ('display.max_columns', None)

#Reading the Dataset
df= pd.read_excel('nba_player_data.xlsx')
```

<h1>Data Cleaning & Analysis Preperation

```python
#Checking null values
df.isna().sum()
```

```python
#Dropping columns that won't be used
df.drop(columns = ['RANK', 'EFF'], inplace=True)
```

```python
df.head(5)
```

```python
#Creating a column that identifies the Year as Integer
df ['season_start_year'] = df ['Year'].str[:4].astype(int)
```

```python
df.head(2)
```

```python
df.TEAM.nunique()
```

```python
df.TEAM.unique() #We have 31 Teams instead of 30, because of the rebranding of the Pelicans
```

```python
#Changing NOP TO NOH. (New Orleans Pelicans ~ Hornets)

df['TEAM'].replace(to_replace= ['NOP','NOH'], value='NO', inplace = True)
```

```python
#Season_Type -> Turning Regular%20Season to RS

df['Season_type'].replace('Regular%20Season', 'RS', inplace= True)
```

```python
df.head(3)
```

```python
#Creating a df for RS and one for the Playoffs

rs_df = df[df['Season_type'] == 'RS']
po_df = df[df['Season_type']=='Playoffs']
```

```python
df.columns
```

```python
#Keeping columns that have no ratios
total_cols = ['MIN','FGM','FGA','FG3M','FG3A','FTM','FTA',
              'OREB','DREB','REB','AST','STL','BLK','TOV','PF','PTS']
```

<h1>Which player stats are correlated with each other?

```python
#Groupping by player and year
df_per_min = df.groupby(['PLAYER','PLAYER_ID','Year'])[total_cols].sum().reset_index()
for col in df_per_min.columns[4:]: #First 4 cols
    df_per_min[col] = df_per_min[col]/df_per_min['MIN']

#Creating the Ratios
df_per_min['FG%'] = df_per_min['FGM']/df_per_min['FGA']
df_per_min['3PT%'] = df_per_min['FG3M']/df_per_min['FG3A']
df_per_min['FT%'] = df_per_min['FTM']/df_per_min['FTA']
df_per_min['FG3A%'] = df_per_min['FG3A']/df_per_min['FGA']
df_per_min['PTS/FGA'] = df_per_min['PTS']/df_per_min['FGA']
df_per_min['FG3M/FGM'] = df_per_min['FG3M']/df_per_min['FGM']
df_per_min['FTA/FGA'] = df_per_min['FTA']/df_per_min['FGA']
df_per_min['TRU%'] = 0.5*df_per_min['PTS']/(df_per_min['FGA']+0.475*df_per_min['FTA']) #Formula for true percentage.(More weight on 3s)
df_per_min['AST_TOV'] = df_per_min['AST']/df_per_min['TOV']
```

```python
#Minimum number of minutes = 50 so that small sample size don't mess the ratios
df_per_min = df_per_min[df_per_min['MIN']>=50]
df_per_min.drop(columns='PLAYER_ID', inplace=True)
```

```python
#First Look on the correlations
df_per_min.corr(numeric_only=True)
```

```python
#Creating a Heatmap for the correlations
fig = px.imshow(df_per_min.corr(numeric_only=True))
fig.show()
```

<h1>How are minutes played distributed?

```python
fig = px.histogram(x=rs_df['MIN'], histnorm='percent', labels={"x": "Minutes played in Regular Season"})
fig.show()
#Shows that many players dont play a lot. 12.47% of players wont play more than 100 minutes in rs.
```

```python
fig = px.histogram(x=po_df['MIN'], histnorm='percent', labels={"x": "Minutes played in Playoffs"})
fig.show()
```

```python
#comparing the 2 graphs by using a complete histogram that includes both
def hist_data(df=rs_df, min_MIN=0,min_GP=0):
    return df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'MIN'] /\
    df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'GP']

fig = go.Figure()
fig.add_trace(go.Histogram(x=hist_data(rs_df,50,5), histnorm='percent', name='RS', xbins={'start':0,'end':46,'size':1})) 
fig.add_trace(go.Histogram(x=hist_data(po_df,5,1), histnorm='percent', name='Playoffs', xbins={'start':0,'end':46,'size':1}))

fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.5)
fig.show()

#Rotation getting tighter in the playoffs
```

```python
((hist_data(rs_df,50,5) >=12) & (hist_data(rs_df,50,5)<=33)).mean()
```

```python
((hist_data(po_df,50,5) >=12) & (hist_data(po_df,50,5)<=33)).mean()
```

```python
#Rotations are getting a lot shorter in PlayOffs
```

```python
#Same for Points
def hist_data(df=rs_df, min_MIN=0,min_GP=0):
    return df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'PTS'] /\
    df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'GP']

fig = go.Figure()
fig.add_trace(go.Histogram(x=hist_data(rs_df,50,5), histnorm='percent', name='RS', xbins={'start':0,'end':38,'size':1})) 
fig.add_trace(go.Histogram(x=hist_data(po_df,5,1), histnorm='percent', name='Playoffs', xbins={'start':0,'end':38,'size':1}))

fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.5)
fig.show()
```

```python
#Similar right-skewed distribution. Shows that star players getting more points on the Playoffs.
```

<h1>How has the game changed over the past 10 years?

```python
change_df= df.groupby('season_start_year')[total_cols].sum().reset_index()

change_df
```

```python
#Estimating no of Possessions
change_df['POSS_est'] = change_df['FGA']-change_df['OREB']+change_df['TOV']+0.44*change_df['FTA']
change_df = change_df[list(change_df.columns[0:2])+['POSS_est']+list(change_df.columns[2:-1])]
change_df
```

```python
#Adding the Ratios

change_df['FG%'] = change_df['FGM']/change_df['FGA']
change_df['3PT%'] = change_df['FG3M']/change_df['FG3A']
change_df['FT%'] = change_df['FTM']/change_df['FTA']
change_df['AST%'] = change_df['AST']/change_df['FGM']
change_df['FG3A%'] = change_df['FG3A']/change_df['FGA']
change_df['PTS/FGA'] = change_df['PTS']/change_df['FGA']
change_df['FG3M/FGM'] = change_df['FG3M']/change_df['FGM']
change_df['FTA/FGA'] = change_df['FTA']/change_df['FGA']
change_df['TRU%'] = 0.5*change_df['PTS']/(change_df['FGA']+0.475*change_df['FTA'])
change_df['AST_TOV'] = change_df['AST']/change_df['TOV']

change_df
```

```python
change_per48_df = change_df.copy() #48 minutes of a match
for col in change_per48_df.columns[2:18]:
    change_per48_df[col] = (change_per48_df[col]/change_per48_df['MIN'])*48*5

change_per48_df.drop(columns='MIN', inplace=True)
change_per48_df
```

```python
fig = go.Figure()
for col in change_per48_df.columns[1:]:
    fig.add_trace(go.Scatter(x=change_per48_df['season_start_year'],
                             y=change_per48_df[col], name=col))
fig.show()
```

```python
#Huge increase on 3PA (From 20 to 35)
#More Efficient scorers. True percentage from 0.53 to 0.56
#More Points getting scored.
```
